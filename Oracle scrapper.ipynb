```python
#importing dependencies
import pandas as pd , numpy as np 
from selenium import webdriver
import colorama
from colorama import Fore
from datetime import datetime

import warnings
warnings.filterwarnings("ignore")

%autosave 1
```

```python
# creating the list of dates 
list_of_dates = []

try:
    #taking input from user
    start_date = input("Enter start date (in YYYY MM DD) : " )
    end_date = input("Enter end date (in YYYY MM DD) : " )
#     end_date = '2022 4 15'
#     start_date = '2000 1 27'

    #adding up condition 
    if start_date > end_date :
        print(Fore.RED+ '\nstart date must come before end date')

    #appending to the list includes both dates
    else:
        for d in pd.date_range(start_date, end_date,inclusive="both"):
            list_of_dates.append(d.strftime("%b %#d, %Y"))

#dealing with error/s    
except ValueError:
    print(Fore.RED +"\nAdd the dates in correct format (in YYYY MM DD)")
```

```python
#driver for operation
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
url = 'https://www.oracle.com/news/'

#getting the url
driver.get(url)

#defining the seed
# if in future it is giving error then try increasing the seed value
seed = 1000

#using try except so that our operation will run smoothly.
try:
    for more in range(seed):
        #clicking on see more button
        driver.find_element_by_xpath('''//*[@id="__next"]/div/div[2]/div/div[1]/section/div/div[2]/div''').click()
        
except:
    pass
```

```python
#letting the operation to sleep so required script will run in the backend 
import time
time.sleep(2)
```

```python
#creating the lists 
titles = []
dates = []
urls = []

try:
    for number in range(1,seed):
        
        #getting all of the dates avilable
        date = driver.find_element_by_xpath(f'''//*[@id="__next"]/div/div[2]/div/div[1]/section/div/div/div[{number}]/div[1]/div''').text
        
        #moving forward only if the date is in list_of dates 
        if date in list_of_dates :
            print(f"found date in news {date}")
            date = driver.find_element_by_xpath(f'''//*[@id="__next"]/div/div[2]/div/div[1]/section/div/div/div[{number}]/div[1]/div''').text
            dates.append(date)
            title = driver.find_element_by_xpath(f'''//*[@id="__next"]/div/div[2]/div/div[1]/section/div/div/div[{number}]/div[2]/h5/a''').text
            titles.append(title)
            href = driver.find_element_by_xpath(f'''//*[@id="__next"]/div/div[2]/div/div[1]/section/div/div/div[{number}]/div[2]/h5/a''').get_attribute('href')
            urls.append(href)
#             print(f"{title}\n{date}\n{href}\n")
except:
    pass
```

```python
#zippling the list so that it'll form the tuple
zipped = list(zip(dates,titles,urls))
zipped[:5]
```

```python
#creating dataframe 
temp_df = pd.DataFrame(zipped,columns=['date','title','url'])
temp_df
```

```python
import re 
import time
time.sleep(3)
```

```python
#creating list for articles
articles =[]

try:
    for x in range(len(temp_df)):
        
        url = temp_df['url'][x]
        
        #getting the url,refreshing the page and letting wait for some time
        driver.get(url)
        driver.refresh()
        time.sleep(3)
        
        #the below loop is able to generate the articles 
        #can mark the invalid links 
        #can report the links which are new
        
        if len(re.findall('https://blogs.oracle.com/',url)) != 0:
            try:
                article = driver.find_element_by_xpath('''/html/body/div/div[2]/div/div/div/div''').text
                articles.append(article)
                print(Fore.GREEN + f"scrapping done for url {x}\n")
            except :
                articles.append("Invalid url")
                print(Fore.RED + f"Invalid url at {x}")
                print(url)
                print()

        elif len(re.findall('http://blogs.oracle.com/',url)) != 0:
            try:
                article = driver.find_element_by_xpath('''//*[@id="ab0f5670-4ae1-4b02-b02b-6d2e71bb01b1"]/div/div''').text
                articles.append(article)
                print(Fore.GREEN + f"scrapping done for url {x}\n")
            except :
                articles.append("Invalid url")
                print(Fore.RED + f"Invalid url at {x}")            
                print(url)
                print()
 
        elif len(re.findall('https://www.oracle.com/emea/news/',url)) != 0:
        
            try:                                          
                article = driver.find_element_by_xpath('''/html/body/div/section[2]/div''').text
                articles.append(article)
                print(Fore.GREEN + f"scrapping done for url {x}\n")
            except :
                articles.append("Invalid url")
                print(Fore.RED + f"Invalid url at {x}")
                print(url)
                print()
                
        elif len(re.findall('https://www.oracle.com/',url)) != 0:
            try:
                article = driver.find_element_by_xpath('''//*[@id="__next"]/div/section[2]/div''').text
                articles.append(article)
                print(Fore.GREEN + f"scrapping done for url {x}\n")
            except :
                articles.append("Invalid url")
                print(Fore.RED + f"Invalid url at {x}")
                print(url)
                print()
                
                
        elif len(re.findall('https://soundcloud.com/',url)) != 0:
            try:
                article = driver.find_element_by_xpath('''//*[@id="content"]/div/div[3]/div[1]/div/div[2]/div[2]/div/div[2]/div/div/div/div[1]/div''').text
                articles.append(article +"\n" + "to load more sign in via link" )
                print(Fore.GREEN + f"scrapping done for url {x}\n")
            except :
                articles.append("Invalid url")
                print(Fore.RED + f"Invalid url at {x}")
                print(url)
                print()
                
        elif len(re.findall('https://investor.oracle.com/',url)) != 0:
            try:
                article = driver.find_element_by_xpath('''//*[@id="_ctrl0_ctl42_divModuleContainer"]/div''').text
                articles.append(article)
                print(Fore.GREEN + f"scrapping done for url {x}\n")
            except :
                articles.append("Invalid url")
                print(Fore.RED + f"Invalid url at {x}")
                print(url)
                print()
                
        elif len(re.findall('https://medium.com/',url)) != 0:
            try:
                article = driver.find_element_by_xpath('''//*[@id="root"]/div/div[3]/div/div/main/div/div[3]/div[1]''').text
                articles.append(article)
                print(Fore.GREEN + f"scrapping done for url {x}\n")
            except :
                articles.append("Invalid url")
                print(Fore.RED + f"Invalid url at {x}")
                print(url)
                print()
                
        elif len(re.findall('https://www.linkedin.com/',url)) != 0:
            try:
                article = driver.find_element_by_xpath('''//*[@id="main-content"]/section[1]/div/article''').text
                articles.append(article)
                print(Fore.GREEN + f"scrapping done for url {x}\n")
            except :
                articles.append("Invalid url")
                print(Fore.RED + f"Invalid url at {x}")
                print(url)
                print()
                
        else:
            print(Fore.BLACK + f"new url found at {x} row")
            print(temp_df['url'][x])
            print()

except:
    pass
```

```python
#adding the output to the dataframe
temp_df['article'] = articles
temp_df
```

```python
#the invalid links are marked as invalid url
#the urls is invalid if either the given url changes after opening or a wrong url
#for example there was a link of the pdf which is identified as invalid url 
temp_df.isna().sum()
```

```python
#saving the csv and keeping the index as false
temp_df.to_csv('Oracle all news release urls.csv',index=False)
```
